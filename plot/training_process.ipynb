```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

configs = {
    "font.size": 20,
    "font.family": "Times New Roman",
}

rcParams.update(configs)

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
model_name = "ST_Transformer"
train_loss = np.load(f"../log/{model_name}/training_log.npy")
val = np.load(f"../log/{model_name}/validation_log.npy")

val_loss = val[:, 1]
print(train_loss.shape, val_loss.shape)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

x_t = np.arange(len(train_loss))
x_v = np.arange(0, len(train_loss), 10)
print(x_t.shape, x_v.shape)
print(train_loss.shape, val_loss.shape)



ax.plot(x_t, train_loss, label="Training Loss")
ax.plot(x_v, val_loss, label="Validation Loss")

ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()

plt.tight_layout()
plt.show()

```
